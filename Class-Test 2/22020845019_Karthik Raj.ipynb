## Explain the Object Oriented Programming concepts   (3)

Object Oriented programming language uses objects which contains data in the form of fields. The different concepts are as follows. 

Abstraction - It is the process of obtaining only the relevant information by removing all the unneccessary data. This is done by hiding the irrelevant data. 

Polymorphism - It is the process where the same object takes different forms or performs different functions. 

Inheritance - It is the process where a new class is derived from an existing class with all its properties.

Encapsulation - It is the process where data is wrapped up into a single unit.

```python
## print factors

def print_factors(x):
    print("The factors of ",x,"are")
    for i in range(1,x+1):
        if x % i==0:
            print (i)
            
num=100

print_factors(num)

## input = 100, output = [1, 2, 4, 5, 10, 20, 25, 50, 100]

## Add comments wherever needed      (3)
```

```python
# Add 45 days to the list and print it as due date (3)
# use functions datetime, strptime, timedelta 

from datetime import datetime
from datetime import timedelta
names_list =[{
    "name": "Joe",
    "issue_date":"30-12-2022"},
    {"name": "John",
    "issue_date":"15-01-2022"}]

for i in names_list:
    new=datetime.strptime(i["issue_date"], "%d-%m-%Y") + timedelta(days=45)
    i["due_date"] = str(new.date())
    print(i)
```
