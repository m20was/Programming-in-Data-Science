# Explain the Object Oriented Programming concepts        (3)

Object oriented programming is based on the concept of "objects" which can contain data and code. It is a programming model 
that organizes software design around data or objects rather than functions and logics. The various concepts of OOP are - classes/objects, encapsulation, inheritance, polymorphism and data abstraction.

encapsulation - it is one of the fundamental concept in object oriented programming(oop). it describes the idea of wrapping data and the methods that work on data within one unit, for eg., a class in java. this concept is often used to hide the internal state representation of an object from the outside.

inheritance - inheritance is the ability of one class to inherit capabilities or properties of another class called the parent class.when we write a class we inherit properties from other classes. so when we create a class we do not need to write all the 
properties and functions again and again as these can be inherited from another class that possesses it. inheritance allows the
user to reuse the code whenever possible and reduce its redundancy.

polymorphism - it is the ability of the data to be processed in more than one form. it allows the performance of the same task in various ways. it consists of method overloading and method overriding i.e. writing the method once and performing number of tasks using the same method name.

data abstraction - it refers to the act of representing the important and special features without including the background details oe explanation about that feature. data abstraction simplifies data design.




```python
# print factors
# input = 100, output = [1, 2, 4, 5, 10, 20, 25, 50, 100]

# Add comments wherever needed      (3)
num =100
factors = []
for i in range(1, num+1):
    if num%i == 0:
        factors.append(i)
print(factors)
```

```python
# Add 45 days to the list and print it as due date (3)
# use functions datetime, strptime, timedelta 
import datetime as dt

from datetime import datetime
from datetime import timedelta
names_list =[{
    "name": "Joe",
    "issue_date":"30-12-2022"},
    {"name": "John",
    "issue_date":"15-01-2022"}]



for i in names_list:
    new=datetime.strptime(i["issue_date"], "%d-%m-%Y")+timedelta(days=45)
    i["due_date"] = str(new.date())
    print(i)
```
