Object oriented programming is a concept that uses objects and classes to depict real world entites. The main concept of OOPs is to combine data, methods and functions into one block of code so that other parts of the code do not access it. Other major concepts of OOPs include: 
(1) Inheritance: capability of one class to imbibe properties from another class
(2) Polymorphism: Multiple forms of the same function
(3) Encapsulation: idea of wrapping data and the methods that work on data within one unit

```python
def fac(num):
    x=[]
    for i in range(1,num+1):
        if num % i == 0:
            x.append(i)
    print(x)
        
            
fac(100)
```

```python
from datetime import datetime
from datetime import timedelta
names_list =[{
    "name": "Joe",
    "issue_date":"30-12-2022"},
    {"name": "John",
    "issue_date":"15-01-2022"}]

for i in names_list:
    new = datetime.strptime(i["issue_date"], "%d-%m-%Y") + timedelta(days=45)
    i["due_date"] = str(new.date())
    print(i)
                       
```
