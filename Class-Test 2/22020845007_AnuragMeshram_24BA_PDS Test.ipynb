# Main concepts of object oriented programming
#1) Class: - This is used to create objects. It contains the basic elements from which an object can be created
#2) object:- An object is any entity which is related to a value. Integers, strings, lists, dictionaries all are objects.
#3) inheritance: - It is the ability of a class to derive the characteristics from another class. The derived class is called child class whereas the class from which characteristics are derived is called parent class.
#4) Polymorphism: - It means the ability to have many forms. The data can be represented in more htan one way.
#5) Encapsulation :- It describes the idea of wrapping data and the methods that work on data within one unit. This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data.
#6)Data Abstraction:- It hides the unnecessary code details from the user. Also,  when we do not want to give out sensitive parts of our code
implementation and this is where data abstraction came.

```python
from datetime import datetime 
from datetime import timedelta
names_list =[{
    "name": "Joe",
    "issue_date":"30-12-2022"},
    {"name": "John",
    "issue_date":"15-01-2022"}]
for i in names_list:
    new=datetime.strptime(i["issue_date"], "%d-%m-%Y")+timedelta(days=45)
    i["due date"]=str(new.date())
    print(i)
```
