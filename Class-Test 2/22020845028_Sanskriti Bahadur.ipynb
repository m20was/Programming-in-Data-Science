```python
# Explain the Object Oriented Programming concepts        (3)

Object Oriented Programming concept (OOPS) is a programming paradigm that uses objects and classes in programming. 
The concept of OOPs is to bind the data and the functions that work on that together as a single unit so that no other part of the code can access the data.
Main concepts of OOPs are -
a. class - A class is a collection of obejects. It is created by keyword class. It contains the blueprints from which the objects are being created.  It is a logical entity that contains attributes and methods
b. The object is an entity that has a state and behavior associated with it.
c. Inheritance - Inheritance is the capability of one to derive or inherit the properties from another class.
It is that derives properties is called the derived class or child class and the class from which the 
properties are being derived is called the base class or parent class.
d. Polymorphism - It simply means having many forms. We can do it using single functions with multiple data. 
e. Encapsulation it is the idea of wrapping data and the methods that work on data within one unit.
```
