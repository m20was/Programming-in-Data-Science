### Explain the Object Oriented Programming concepts     (3)
Object oriented programming is a computer programming model that organizes software design around data, or objects, rather than functions and logic. 

<h1> 5 Basic Concepts of object-oriented programming are: </h1>

1)Classes and Objects: A class is called a Maple (capital M). Because it is the blueprint, we capitalise it. The growing maple tree is actually an instance, or a type, of the Maple class when seeds fall and sprout. The maple tree that is currently growing is regarded as an object, or an instance of a class.

2)Inheritance: Remember that a spruce tree will contain characteristics of all trees in addition to characteristics of evergreen trees. All of the attributes of the Tree class will be inherited by a Spruce class. Classes in object-oriented design have parents much like our DNA and traits do. In this instance, the Spruce class has access to the variables and methods in the Tree class.

3)Encapsulation : The shielding, or concealment, of data is referred to as encapsulation. The advantage of object-oriented design is that you can prevent the use of specific data in specific classes. Keep in mind that everything from the Evergreen class was passed down to the Spruce class.

4)Abstraction :This is the fundamental idea of object-oriented programming, which enables programmers to abstract or isolate recurring traits in objects and methods. In other words, hiding internal details and displaying functionality are both examples of abstraction. Abstraction is mostly used by programmers to manage complexity by obscuring unimportant elements. This builds on encapsulation.

5)Polymorphism : The fundamental idea of object-oriented programming is polymorphism, which enables developers to create logical codes.
