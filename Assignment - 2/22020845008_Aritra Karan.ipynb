# Prime Number between two numbers
# Create a function prime_factors
# prime_factors(2,19) = [2, 3, 5, 7, 11, 13, 17, 19]

```python
def prime_factors(num1, num2):
    prime_num = []
    for num in range(num1, num2+1):
        div = []
        for i in range(1,num+1):
            if num%i == 0:
                div.append(i)      
        if len(div) == 2:
            prime_num.append(num)
        else:
            pass
    print(prime_num)
      
prime_factors(3,36) 
```

#Q.3
#Calculate Factorial of a number
# factorial 5 : 5*4*3*2*1  = 120
# Recursive function

```python
def factorial(num):
    if num == 1 or num == 0:
        return 1
    else:
        return num * factorial(num-1)

print(factorial(8))
```

#Q.4
# Python program to check if year is a leap year or not
# Define a function is_leap_year and call it
# Pass the input in an array and call the function with a for loop

```python
def is_leap_year(year):
    if (year % 4) == 0:
        if (year % 100) == 0:
            if (year % 400) == 0:
                print('leap')
            else:
                print('no leap')
        else:
            print('leap')
    else:
        print('no leap')

data = [2011, 1998, 4096, 2000, 2022]

for i in range(len(data)):
    is_leap_year(data[i])
```

#Q.5
# Compute HCF
# compute_hcf(320, 132)


```python
def compute_hcf(a, b):
    if(b == 0):
        return abs(a)
    else:
        return compute_hcf(b, a % b)
        
print(compute_hcf(167, 876))
    
```
