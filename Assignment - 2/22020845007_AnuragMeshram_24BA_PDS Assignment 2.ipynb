```python
#Q.1
# Prime Number between two numbers
# Create a function prime_factors
# prime_factors(2,19) = [2, 3, 5, 7, 11, 13, 17, 19]|
```

```python
#prime numbers between 2 numbers
a=2
b=20
def prime(x, y):
    prime_list = []
    for i in range(x, y):
        if i == 0 or i == 1:
            continue
        else:
            for j in range(2, int(i/2)+1):
                if i % j == 0:
                    break
            else:
                prime_list.append(i)
    return prime_list
prime(a,b)
 
```
