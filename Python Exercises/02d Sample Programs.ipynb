```python
# Python program to check if year is a leap year or not

year = 1900
is_leapyr = False
if year%400 == 0: 
    is_leapyr = True
elif year%100 == 0: 
    is_leapyr = False
elif year%4 == 0:
    is_leapyr = True
print(f"year {year} - {is_leapyr}" )
```

```python
def is_leap(year):
    is_leapyr = False
    if year%400 == 0: 
        is_leapyr = True
    elif year%100 == 0: 
        is_leapyr = False
    elif year%4 == 0:
        is_leapyr = True
    return is_leapyr

year = 1900
print(f"{year} - {is_leap(year)}")
```

```python
#Factorial computation
def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*fact(n-1)

fact(4)
```
