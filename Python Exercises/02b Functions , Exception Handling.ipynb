## Function

<h4>
A function is a block of code which only runs when it is called. <br>
You can pass data, known as parameters, into a function. <br>
A function can return data as a result. <br>
Arguments, Parameters : An argument is the value that is sent to the function when it is called. <br>
</h4>

```python
#Global Variables, Defines the scope of a variable
yy = "yes"
def myfunc():
    global yy
    yy = "awesome"
    print("Python is " + yy)
    

myfunc()
print(yy)
```

```python
# Passing arguments to a function
def my_function(first_name, last_name="no last name"):
  print(first_name + " " + last_name)

my_function("Emily", "George")

my_function(last_name="George",first_name = "Emily")
my_function("Amy")
```

```python
#Keyword argument , You can also send arguments with the key = value syntax.
#You can also send arguments with the key = value syntax.

def my_function(child1="1",child2="a",child3="d"):
    print("The youngest child is ", child3)
    
my_function(child3="Emil", child2 ="Tobias",child1 = "Linus")
my_function("a","b")
my_function()
```

```python
#Arbitrary Arguments, *args
#If the number of arguments is unknown, add a * before the parameter name:
def my_function(*kids):
    
    arg_len = len(kids)
    if arg_len > 0:
        print("The youngest child is " + kids[arg_len-1])
    else:
        print("No names passed")
    
my_function("Emil", "Tobias","Lucy")
my_function()
```

```python
# Arbitrary Keyword Arguments, **kwargs
def my_function(**kid):
# Function expects values in lname, fname, mname as strings 
  print("His last name is ", kid.get("lname"))
  print("His middlename is ",kid.get("mname"))

my_function(fname = "Tom",lname = "Jackson")
```

```python
#Lambda function
x = lambda a : a + 10
print(x(5))

x = lambda a, b : a * b
print(x(5, 6))

x = lambda a, b, c : a + b + c
print(x(5, 6, 2))


```

```python
# Use of lambda function
def myfunc(n):
  return lambda a : a * n

table_2 = myfunc(2) ## lambda a : a * 2
table_3 = myfunc(3) ## lambda a : a * 3

for i in range(10):
    print(table_2(i))

print(table_3(11))
```
