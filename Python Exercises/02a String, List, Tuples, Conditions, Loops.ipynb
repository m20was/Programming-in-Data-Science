## Sets

```python
# Set operations

this_set = {"apple", "banana", "cherry", "apple"} # drops duplicates
print(this_set)

this_set = set(("apple", "banana", "cherry","orange"))

print(this_set)
this_set.add("mango")

print(this_set)
print("banana" in this_set)

this_set.remove("banana")
print(this_set)
```

```python
# Join sets
set1 = {"a", "b" , "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)
print(set3)
```

### Dictionary

```python
#Dictionary 
x = {
     "name" : "John", 
     "age" : 36
    }
print(x,' ',type(x))
```

```python
# Dictionary


vehicle_details = {
  "brand": "Ford",
  "model": "Aspire",
  "year": 1997
}
# Retrieve value
x = vehicle_details["model"]
print(x)

x = vehicle_details.get("model")
print(x)

# update a key value
print("Before",vehicle_details.get("year"))

vehicle_details["year"] = 2018
print("After",vehicle_details.get("year"))

vehicle_details.update({"year": 2020})
print("After",vehicle_details.get("year"))

# add a new key
vehicle_details.update({"color": "red"}) # add a new key
print(vehicle_details)
```

```python
for key, value in vehicle_details.items():
  print(key, value)
```

```python
# Looping a dictionary

for x in vehicle_details:
  print(x,vehicle_details[x])
print("------------")


for x in vehicle_details.values():
  print(x)
print("------------")

for x in vehicle_details.keys():
  print(x)
print("------------")

for key, value in vehicle_details.items():
  print(key, value)
```

```python
# for loops

fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

```python
for x in range(2, 10): #Increment the sequence with 3 (default is 1):
  print(x)
print('--------')
```

```python
for x in range(2, 16, 3): #Increment the sequence with 3 (default is 1):
  print(x)
print('--------')


for x in range(3):
  print(x)
else:
  print("Finally finished!")
```

```python
# Nested Loops
adj = ["red", "big", "tasty"]
fruits = ["apple", "orange", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```

### List Comprehensions

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist = [len(x) for x in fruits]
print(newlist)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist = [x for x in fruits if "a" in x]
print(newlist)

newlist = [x for x in fruits if x != "apple"]
print(newlist)

newlist = [x for x in range(10)]
print(newlist)

newlist = [x for x in range(10) if x % 2 ==0]
print(newlist)

newlist = ['hello' for x in fruits]
print(newlist)
```
